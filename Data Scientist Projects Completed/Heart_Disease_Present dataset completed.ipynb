# Data preprocessing

```python
from sklearn.preprocessing import MinMaxScaler
mms=MinMaxScaler()
df[["resting_blood_pressure","chest_pain_type","num_major_vessels","fasting_blood_sugar_gt_120_mg_per_dl",
      "serum_cholesterol_mg_per_dl","oldpeak_eq_st_depression"]]=\
mms.fit_transform(df[["resting_blood_pressure","chest_pain_type","num_major_vessels","fasting_blood_sugar_gt_120_mg_per_dl",
      "serum_cholesterol_mg_per_dl","oldpeak_eq_st_depression"]])
```

```python
#It is the specific outlier detection
q1=np.percentile(df['resting_blood_pressure'],25)
q3=np.percentile(df['resting_blood_pressure'],75)
IQR=q3-q1
upper=q3+(1.5*IQR)
lower=q1-(1.5*IQR)
print(lower,q1,IQR,q3,upper)
print(max(df.resting_blood_pressure),min(df.resting_blood_pressure))
```

```python
olst=["resting_blood_pressure","chest_pain_type","num_major_vessels","fasting_blood_sugar_gt_120_mg_per_dl",
      "serum_cholesterol_mg_per_dl","oldpeak_eq_st_depression"] #for loop outliers detection all
for a in olst:
    q1=np.percentile(df[a],25)
    q3=np.percentile(df[a],75)
    iqr=q3+(1.5*IQR)
    upper=q3+(1.5*IQR)
    median=np.median(df[a])
    print(a)
    print(lower,median,upper)
    print("Above:",df.loc[df[a]>upper].count()[0])
    print("Below:",df.loc[df[a]<lower].count()[0])
    if df.loc[df[a]>upper].count()[0]>0:
        df.loc[df[a]>upper,a]=median
    if df.loc[df[a]<lower].count()[0]>0:
        df.loc[df[a]<lower,a]=median
```

```python
plt.figure(figsize=(15,20))
plotnumber = 1
for i in df:
    plt.subplot(7,3,plotnumber)
    sns.boxplot(x=i,data=df)
    plotnumber+=1
```

# Decision Tree Classifier

```python
from imblearn.over_sampling import SMOTE
smote=SMOTE()
xtrain_sm,ytrain_sm=smote.fit_resample(xtrain,ytrain)
```

```python
from collections import Counter
print("before SMOTE:",Counter(ytrain))
print("After SMOTE:",Counter(ytrain_sm))
```

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier(random_state=10)
dt.fit(xtrain_sm,ytrain_sm)
```

```python
ypred=dt.predict(xtest)
```
